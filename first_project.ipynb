```python
# –ü–æ—Å–º–æ—Ç—Ä–∏–º —Å—Ä–µ–¥–Ω–µ–µ, –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ
data['ceiling_height'].describe()
```

<span style="color:firebrick">**–ü—Ä–æ–±–ª–µ–º–∞:** </span> –í —Å—Ç–æ–ª–±—Ü–µ `ceiling_height` –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –¥–æ—Å—Ç–∏–≥–∞–µ—Ç 100 –º., –∞ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ 1 –º., —Å–æ–º–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –≤—ã—Å–æ—Ç—ã –¥–ª—è –ø–æ—Ç–æ–ª–∫–æ–≤. –ü–æ—Å–º–æ—Ç—Ä–∏–º –∫–∞–∫–∏–µ –∑–∞–Ω—á–µ–Ω–∏—è –µ—â–µ –∏–º–µ—é—Ç—Å—è –≤ —Ç–∞–±–ª–∏—Ü–µ.

```python
# –í—ã–≤–æ–¥–∏–º —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Å—Ç–æ–ª–±—Ü–µ ceiling_height
data['ceiling_height'].unique()
```

–í —Ç–∞–±–ª–∏—Ü–µ –ø—Ä–∏—Å—É—Å—Ç–≤—É—é—Ç –∑–Ω–∞—á–µ–Ω–∏—è –±–æ–ª–µ–µ 8 –º. –¢–∞–∫–∏–µ —á–∏—Å–ª–∞, –≤–æ–∑–º–æ–∂–Ω—ã –∏–∑-–∑–∞ –æ—à–∏–±–∫–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –ø—Ä–∏ –≤–≤–æ–¥–µ. –ó–Ω–∞—á–µ–Ω–∏—è >10 –ø–æ–¥–µ–ª–∏–º –Ω–∞ 10, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ, —á—Ç–æ–±—ã –ø—Ä–∏–≤–µ—Å—Ç–∏ –∫ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–º—É –≤–∏–¥—É. –ó–Ω–∞—á–µ–Ω–∏—è –º–µ–Ω—å—à–µ 2 –∑–∞–º–µ–Ω–∏–º –Ω–∞ NoneType, —Ç.–∫., —É–¥–∞–ª—è—è —ç—Ç–∏ —Å—Ç—Ä–æ–∫–∏, –º—ã —Ç–µ—Ä—è–µ–º –¥–∞–Ω–Ω—ã–µ. –¢–∞–∫ –∂–µ, –∑–∞–º–µ–Ω–∏–º –ø—Ä–æ–ø—É—Å–∫–∏ –º–µ–¥–∏–∞–Ω–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º.

–ó–Ω–∞—á–µ–Ω–∏–π –±–æ–ª—å—à–µ 8 –≤—Å–µ–≥–æ 2, –æ–Ω–∏ –Ω–µ –ø–æ–≤–ª–∏—è—é—Ç –Ω–∞ –∏—Ç–æ–≥–æ–≤—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç, –º–æ–∂–µ–º –Ω–µ –æ–±—Ä–∞—â–∞—Ç—å –Ω–∞ –Ω–∏—Ö –≤–Ω–∏–º–∞–Ω–∏–µ.

```python
# –ò–∑–º–µ–Ω–∏–º —Å—Ç—Ä–æ–∫–∏ —Å –≤—ã—Å–æ—Ç–æ–π –ø–æ—Ç–æ–ª–∫–∞ –±–æ–ª–µ–µ 10 –º. 
data.loc[data['ceiling_height'] > 10, 'ceiling_height'] = data['ceiling_height']/10 

# –ó–∞–º–µ–Ω–∏–º —Å—Ç—Ä–æ–∫–∏ —Å –≤—ã—Å–æ—Ç–æ–π –ø–æ—Ç–æ–ª–∫–∞ –º–µ–Ω–µ–µ 2 –º. –Ω–∞ NoneType
data.loc[data['ceiling_height'] < 2, 'ceiling_height'] = None
```

```python
# –ó–∞–ø–æ–ª–Ω—è–µ–º –ø—Ä–æ–ø—É—Å–∫–∏ –º–µ–¥–∏–∞–Ω–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
```

<a name="2.3.2."></a>**2. –°—Ç–æ–ª–±—Ü—ã foor_total –∏ locality_name:**<br/> 

```python
# –û—Ç–±—Ä–æ—Å–∏–º —Å—Ç—Ä–æ–∫–∏ —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏ –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö floors_total –∏ locality_name
data = data.dropna(subset=['floors_total','locality_name'])

# –ü—Ä–æ–≤–µ—Ä—è–µ–º –æ—Ç—Å—É—Å—Ç–≤—É—é—Ç –ª–∏ –ø—Ä–æ–ø—É—Å–∫–∏ —Ç–µ–ø–µ—Ä—å
data[['floors_total','locality_name']].isna().sum()
```

<a name="2.3.3."></a>**3. –°—Ç–æ–ª–±–µ—Ü balcony:**<br/>

<a name="2.3.4."></a>**4. –°—Ç–æ–ª–±—Ü—ã living_area –∏ kitchen_area:**<br/>

```python
# –°–º–æ—Ç—Ä–∏–º —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–∞ locality_name
data.locality_name.sort_values().unique()
```

<b>–û–±–Ω–∞—Ä—É–∂–µ–Ω–Ω—ã–µ –Ω–µ—è–≤–Ω—ã–µ –¥—É–±–ª–∏–∫–∞—Ç—ã:</b>

- –¢–∞–∫ –∂–µ, —Ä–∞–∑–Ω–æ–µ –Ω–∞–ø–∏—Å–∞–Ω–∏–µ –Ω–∞—Å–µ–ª–µ–Ω–Ω–æ–≥–æ –ø—É–Ω–∫—Ç–∞, –¥–æ–ø—É—Å—Ç–∏–º, `'–ø–æ—Å–µ–ª–æ–∫ –≥–æ—Ä–æ–¥—Å–∫–æ–≥–æ —Ç–∏–ø–∞'` –∏ `'–≥–æ—Ä–æ–¥—Å–∫–æ–π –ø–æ—Å–µ–ª–æ–∫'`.

- –í—Å—Ç—Ä–µ—á–∞—é—Ç—Å—è –ø–æ–≤—Ç–æ—Ä—ã, –Ω–∞–ø—Ä–∏–º–µ—Ä, `'–ö—É–¥—Ä–æ–≤–æ'` –∏ `'–¥–µ—Ä–µ–≤–Ω—è –ö—É–¥—Ä–æ–≤–æ'`.

<b>–†–µ—à–µ–Ω–∏–µ.</b> 

1. –í –ø–µ—Ä–≤–æ–º —Å–ª—É—á–∞–µ, —Å—Ç–æ–∏—Ç –ø—Ä–æ—Å—Ç–æ —Å–¥–µ–ª–∞—Ç—å –∑–∞–º–µ–Ω—É —Å–ª–æ–≤. 

2. –í–æ –≤—Ç–æ—Ä–æ–º, –¥–ª—è –Ω–∞—á–∞–ª–∞ —Å–¥–µ–ª–∞—Ç—å –∫–æ–ø–∏—é —Ç–∞–±–ª–∏—Ü—ã —Å "–¥–ª–∏–Ω—ã–º–∏ –Ω–∞–∑–≤–∞–Ω–∏—è–º–∏", –¥–∞–ª–µ–µ —É –æ—Å–Ω–æ–≤–Ω–æ–π —Ç–∞–±–ª–∏—Ü—ã `data` —É–¥–∞–ª–∏—Ç—å –≤—Å–µ —Ç–∏–ø—ã –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–π –Ω–∞—Å–µ–ª–µ–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–æ–≤, —Ç.–µ. –≤–º–µ—Å—Ç–æ `"–ø–æ—Å–µ–ª–æ–∫ –≥–æ—Ä–æ–¥—Å–∫–æ–≥–æ —Ç–∏–ø–∞ –î—É–±—Ä–æ–≤–∫–∞"` –æ—Å—Ç–∞–Ω–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ –Ω–∞–∑–≤–∞–Ω–∏–µ `"–î—É–±—Ä–æ–≤–∫–∞"`. –ü–æ—Ç–æ–º —Å–æ–∑–¥–∞–µ–º —Å–ª–æ–≤–∞—Ä—å, –≤ –∫–æ—Ç–æ—Ä–æ–º –±—É–¥–µ—Ç —Ö—Ä–∞–Ω–∏—Ç—Å—è –Ω–∞–∑–≤–∞–Ω–∏–µ –∏–∑ –æ—Å–Ω–æ–≤–Ω–æ–π —Ç–∞–±–ª–∏—Ü—ã `data` (–∑–Ω–∞—á–µ–Ω–∏–µ, –∫–æ—Ç–æ—Ä–æ–µ —Ö–æ—Ç–∏–º –∑–∞–º–µ–Ω–∏—Ç—å, —Ç–∞ –∂–µ "–î—É–±—Ä–æ–≤–∫–∞") –∏ –Ω–æ–≤–æ–µ –Ω–∞–∑–≤–∞–Ω–∏–µ –∏–∑ —Ç–∞–±–ª–∏—Ü—ã-–∫–æ–ø–∏–∏ (–∑–Ω–∞—á–µ–Ω–∏–µ, –Ω–∞ –∫–æ—Ç–æ—Ä–æ–µ –∑–∞–º–µ–Ω—è–µ–º, "–ø–æ—Å–µ–ª–æ–∫ –≥–æ—Ä–æ–¥—Å–∫–æ–≥–æ —Ç–∏–ø–∞ –î—É–±—Ä–æ–≤–∫–∞").

<b>–ò—Ç–æ–≥–æ –∞–ª–≥–æ—Ä–∏—Ç–º –≤—Ç–æ—Ä–æ–≥–æ —Å–ª—É—á–∞—è:</b> 
1. –°–æ–∑–¥–∞–µ–º –∫–æ–ø–∏—é-—Ç–∞–±–ª–∏—Ü—ã data <b> (copy_table)</b> , –≤ —Å—Ç–æ–ª–±—Ü–µ locality_name –±—É–¥—É—Ç —Ö—Ä–∞–Ω–∏—Ç—Å—è `"–¥–ª–∏–Ω–Ω—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è"`.

2. –£ –æ—Å–Ω–æ–≤–Ω–æ–π —Ç–∞–±–ª–∏—Ü—ã <b>data</b> —É–¥–∞–ª—è–µ–º –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏—è –Ω–∞—Å–µ–ª–µ–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–æ–≤ `("–ø–æ—Å–µ–ª–æ–∫ –≥–æ—Ä–æ–¥—Å–∫–æ–≥–æ —Ç–∏–ø–∞ –î—É–±—Ä–æ–≤–∫–∞" = "–î—É–±—Ä–æ–≤–∫–∞")`.
3. –°–æ–∑–¥–∞–µ–º –ø—É—Å—Ç–æ–π —Å–ª–æ–≤–∞—Ä—å <b>(dict_name_city)</b>, –∑–∞–ø–æ–ª–Ω—è–µ–º –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ <b> {'—Å—Ç–∞—Ä–æ–µ –Ω–∞–∑–≤–∞–Ω–∏–µ': '–Ω–æ–≤–æ–µ –Ω–∞–∑–≤–∞–Ω–∏–µ'}</b> .
4. –° –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ <b>map()</b> –∑–∞–º–µ–Ω—è–µ–º –≤ —Ç–∞–±–ª–∏—Ü–µ data, —Å—Ç–æ–ª–±—Ü–µ locality_name –Ω–∞ –Ω–æ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∏–∑ —Å–ª–æ–≤–∞—Ä—è <b>dict_name_city`</b>.

```python
# 1) –ó–∞–º–µ–Ω–∏–º –±—É–∫–≤—É '—ë' –≤ —Å–ª–æ–≤–µ '–ø–æ—Å—ë–ª–æ–∫' –Ω–∞ '–µ'
# 2) –ó–∞–º–µ–Ω–∏–º —Ñ—Ä–∞–∑—É '–≥–æ—Ä–æ–¥—Å–∫–æ–π –ø–æ—Å–µ–ª–æ–∫' –Ω–∞ '–ø–æ—Å–µ–ª–æ–∫ –≥–æ—Ä–æ–¥—Å–∫–æ–≥–æ —Ç–∏–ø–∞'
# 3) –ó–∞–º–µ–Ω–∏–º —Ñ—Ä–∞–∑—É '–ø–æ—Å—ë–ª–æ–∫ —Å—Ç–∞–Ω—Ü–∏–∏' –Ω–∞ '–ø–æ—Å—ë–ª–æ–∫ –ø—Ä–∏ –∂–µ–ª–µ–∑–Ω–æ–¥–æ—Ä–æ–∂–Ω–æ–π —Å—Ç–∞–Ω—Ü–∏–∏'
data['locality_name'] = (
    data['locality_name']
    .str.replace('—ë', '–µ')
    .replace(['–≥–æ—Ä–æ–¥—Å–∫–æ–π –ø–æ—Å–µ–ª–æ–∫', '–ø–æ—Å–µ–ª–æ–∫ —Å—Ç–∞–Ω—Ü–∏–∏'], 
             ['–ø–æ—Å–µ–ª–æ–∫ –≥–æ—Ä–æ–¥—Å–∫–æ–≥–æ —Ç–∏–ø–∞', '–ø–æ—Å–µ–ª–æ–∫ –ø—Ä–∏ –∂–µ–ª–µ–∑–Ω–æ–¥–æ—Ä–æ–∂–Ω–æ–π —Å—Ç–∞–Ω—Ü–∏–∏'], regex=True)
)
data.locality_name.sort_values().unique()
```

```python
# –ü—Ä–æ–≤–µ—Ä—è–µ–º –∫–æ–ª-–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ—Å–ª–µ –ø–µ—Ä–≤–æ–π –∑–∞—á–∏—Å—Ç–∫–∏
len(data['locality_name'].unique())
```

<b>–ó–∞—á–∏—â–∞–µ–º –¥—É–±–ª–∏–∫–∞—Ç—ã –ø–æ —Ç–∏–ø—É "–ö—É–¥—Ä–æ–≤–æ" –∏ "–¥–µ—Ä–µ–≤–Ω—è –ö—É–¥—Ä–æ–≤–æ"</b>:

```python
# –®–∞–≥ 1. –£–¥–∞–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏–π NaN, –∏–Ω–∞—á–µ —Ä–∞–±–æ—Ç–∞—Ç—å –æ—Ç–∫–∞–∑—ã–≤–∞–µ—Ç—Å—è
data = data.dropna(subset=['locality_name']).reset_index(drop=True)
```

```python
# –®–∞–≥ 2. –î–µ–ª–∞–µ–º –∫–æ–ø–∏—é data, –≤ –Ω–µ–π —Ö—Ä–∞–Ω–∏–º "–¥–ª–∏–Ω—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è –Ω–∞—Å–µ–ª–µ–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–æ–≤", –¥–æ–ø—É—Å—Ç–∏–º, –≤ data –ø–æ–∑–∂–µ –±—É–¥–µ—Ç —Ö—Ä–∞–Ω–∏—Ç—Å—è "–î—É–±—Ä–æ–≤–∫–∞",
# –∞ –≤ –∫–æ–ø–∏–∏ "–ø–æ—Å–µ–ª–æ–∫ –≥–æ—Ä–æ–¥—Å–∫–æ–≥–æ —Ç–∏–ø–∞ –î—É–±—Ä–æ–≤–∫–∞"
copy_table = data.copy()

# –ü—Ä–æ–≤–µ—Ä–∫–∞ —Ö—Ä–∞–Ω—è—Ç—Å—è –ª–∏ –≤ —Å—Ç–æ–ª–±—Ü–µ "–¥–ª–∏–Ω—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è", —Ç.–µ. ('–ø–æ—Å–µ–ª–æ–∫', '–¥–µ—Ä–µ–≤–Ω—è' –∏ —Ç.–¥.)
copy_table['locality_name'].value_counts().head(10)
```

```python
# –®–∞–≥ 3. –£–¥–∞–ª—è–µ–º —É –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏—è –Ω–∞—Å–µ–ª–µ–Ω—ã—Ö –ø—É–Ω–∫—Ç–æ–≤.
data['locality_name'] = data['locality_name'].apply(
                       lambda title: ' '.join(word for word in title.split(' ') if not(word.islower()))
)

# –°–º–æ—Ç—Ä–∏–º, –∫–∞–∫ –∏–∑–º–µ–Ω–∏–ª—Å—è —Å—Ç–æ–ª–±–µ—Ü
data['locality_name'].value_counts().head(10)
```

```python
# –®–∞–≥ 4. –°–æ–∑–¥–∞–µ–º –ø—É—Å—Ç–æ–π —Å–ª–æ–≤–∞—Ä—å
dict_name_city = dict()
```

```python
# –®–∞–≥ 5. –ø–µ—Ä–µ–±–∏—Ä–∞–µ–º —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è data
for name in data['locality_name'].unique():
    # –°–æ—Ö—Ä–∞–Ω—è–µ–º –∏–Ω–¥–µ–∫—Å —Å—Ç—Ä–æ–∫–∏ copy_table, –µ—Å–ª–∏ –Ω–∞–∑–≤–∞–Ω–∏–µ (name), –¥–æ–ø—É—Å—Ç–∏–º "–ö—É–¥—Ä–æ–≤–æ" —É–ø–æ–º–∏–Ω–∞–µ—Ç—Å—è –≤ —Å—Ç—Ä–æ–∫–µ.
    # –ò–Ω–∞—á–µ –≥–æ–≤–æ—Ä—è, name("–ö—É–¥—Ä–æ–≤–æ") in copy_table("–¥–µ—Ä–µ–≤–Ω—è –ö—É–¥—Ä–æ–≤–æ") == –≤—Ö–æ–∂–¥–µ–Ω–∏–µ –µ—Å—Ç—å, —Å–æ—Ö—Ä–∞–Ω—è–µ–º –∏–Ω–¥–µ–∫—Å.
    indx = np.where(copy_table['locality_name'].str.contains(name, regex=True))
    
    # –î–æ–±–∞–≤–ª—è–µ–º –≤ —Å–ª–æ–≤–∞—Ä—å —Å—Ç–∞—Ä–æ–µ –∏ –Ω–æ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ
    dict_name_city[name] = copy_table['locality_name'].loc[indx[0][0]]
    
# –°–º–æ—Ç—Ä–∏–º, –∫–∞–∫ –∑–∞–ø–æ–ª–Ω–∏–ª—Å—è —Å–ª–æ–≤–∞—Ä—å
dict_name_city
```

```python
# –®–∞–≥ 6. –ò—Å–ø–æ–ª—å–∑—É–µ–º –º–µ—Ç–æ–¥ map() –∏ —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –∑–∞–º–µ–Ω—ã —Å—Ç–∞—Ä—ã—Ö (–∫–æ—Ä–æ—Ç–∫–∏—Ö) –∑–Ω–∞—á–µ–Ω–∏–π –Ω–∞ –Ω–æ–≤—ã–µ (–¥–ª–∏–Ω—ã–µ)
data['locality_name'] = data['locality_name'].map(dict_name_city)

# –°–º–æ—Ç—Ä–∏–º, –∫–∞–∫ –∏–∑–º–µ–Ω–∏–ª—Å—è —Å–ª–æ–ª–±–µ—Ü
data['locality_name'].value_counts().head(10)
```

```python
# –ü—Ä–æ–≤–µ—Ä—è–µ–º –∫–æ–ª-–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ—Å–ª–µ –≤—Ç–æ—Ä–æ–π –∑–∞—á–∏—Å—Ç–∫–∏
len(data['locality_name'].unique())
```

```python
# –°–æ–∑–¥–∞–µ–º —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –æ–ø—Ä–µ–¥–µ–ª–∏—Ç —Ç–∏–ø —ç—Ç–∞–∂–∞
def floor_type(row):
    if row['floor'] == 1:
        return '–ø–µ—Ä–≤—ã–π'
    
    elif row['floor'] == row.floors_total:
        return '–ø–æ—Å–ª–µ–¥–Ω–∏–π'
    
    else:
        return '–¥—Ä—É–≥–æ–π'
```

<a name="4.4."></a>**4.4. –§–∞–∫—Ç–æ—Ä—ã, –≤–ª–∏—è—é—â–∏–µ –Ω–∞ —Å—Ç–æ–∏–º–æ—Å—Ç—å –æ–±—ä–µ–∫—Ç–∞.**<br/> 

–°–æ–∑–¥–∞–¥–∏–º –¥–≤–µ —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è –¥–∏–∞–≥—Ä–∞–º–º:

```python
# –§—É–Ω–∫—Ü–∏—è, —Å–æ–∑–¥–∞—é—â–∞—è –¥–∏–∞–≥—Ä–∞–º–º—É —Ä–∞—Å—Å–µ—è–Ω–∏—è
def scattering_graph(data, column_x, column_y):
    # –î–æ–±–∞–≤–∏–º –∫ –∑–Ω–∞—á–µ–Ω–∏—è–º x, y –Ω–µ–∫–æ—Ç–æ—Ä—ã–π —Å–ª—É—á–∞–π–Ω—ã–π —à—É–º (–¥—Ä–æ–∂–∞–Ω–∏–µ)
    noise = np.random.normal(0, 2, size=len(good_data))
 
    x_jitter = data[column_x] + noise
    y_jitter = data[column_y] + noise
    
    # –û—Ç–æ–±—Ä–∞–∂–∞–µ–º –¥–∏–∞–≥—Ä–∞–º–º—É
    plt.plot(x_jitter, y_jitter, 'o',
             alpha=0.02, markersize=1)
```

```python
# –§—É–Ω–∫—Ü–∏—è, —Å–æ–∑–¥–∞—é—â–∞—è –ª–∏–Ω–µ–π–Ω—É—é —Ä–µ–≥—Ä–µ—Å—Å–∏—é
def linear_regression(data, column_x, column_y):
    # linregress –Ω–µ –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏—è NaN, —É–¥–∞–ª—è–µ–º —Å—Ç—Ä–æ–∫–∏ —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏
    subset = data.dropna(subset=[column_x, column_y])
    
    x_clean = subset[column_x]
    y_clean = subset[column_y]
    
    # –í—ã—á–∏—Å–ª—è–µ–º –ª–∏–Ω–µ–π–Ω—É—é —Ä–µ–≥—Ä–µ—Å—Å–∏—é 
    res_x = linregress(x_clean, y_clean)
    
    # –í—ã—á–∏—Å–ª—è–µ–º –ª–∏–Ω–∏—é —Ç—Ä–µ–Ω–¥–∞
    fx = np.array([x_clean.min(), x_clean.max()])
    fy = res_x.intercept + res_x.slope * fx
    
    # –û—Ç–æ–±—Ä–∞–∂–∞–µ–º
    plt.plot(fx, fy, '-')
```

<a name="4.4.1."></a>**4.4.1. –ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å —Ü–µ–Ω—ã –æ—Ç –æ–±—â–µ–π –ø–ª–æ—â–∞–¥–∏.**<br/> 

```python
# –°–æ–∑–¥–∞–µ–º –¥–∏–∞–≥—Ä–∞–º–º—É —Ä–∞—Å—Å–µ—è–Ω–∏—è –¥–ª—è —Å—Ç–æ–ª–±—Ü–∞ total_area
scattering_graph(good_data, 'total_area', 'last_price')
linear_regression(good_data, 'total_area', 'last_price')

plt.xlim([0, 125])
plt.ylim([0, 3e7])

plt.xlabel('–û–±—â–∞—è –ø–ª–æ—â–∞–¥—å (–º¬≤)', size=14)
plt.ylabel('1 –µ–¥. = ( 5 –º–ª–Ω.—Ä—É–±.)', size=14)
plt.title('–ó–∞–≤–∏—Å–∏—Å–º–æ—Å—Ç—å —Ü–µ–Ω—ã –æ—Ç –æ–±—â–µ–π –ø–ª–æ—â–∞–¥–∏', size=14)
```
